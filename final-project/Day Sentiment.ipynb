```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('clean_tweets.csv') 
```

# New df containing just sentiments

```python
new_df = data[['retweet_count', 'created_at','sentiment']].copy()
```

```python
new_df
```

```python

new_df[['positive', 'neutral', 'negative']] = pd.DataFrame([[0,0,0]])
```

```python
new_df
```

```python
def fillPositive(new_df):
    if new_df['sentiment'] == 'positive':
        value = new_df['retweet_count']+1
        return value
    else:
        return 0
```

```python
def fillNeutral(new_df):
    if new_df['sentiment'] == 'neutral':
        value = new_df['retweet_count']+1
        return value
    else:
        return 0
```

```python
def fillNegative(new_df):
    if new_df['sentiment'] == 'negative':
        value = new_df['retweet_count']+1
        return value
    else:
        return 0
```

```python
new_df['positive'] = new_df.apply(fillPositive, axis=1)
```

```python
new_df['neutral'] = new_df.apply(fillNeutral, axis=1)
```

```python
new_df['negative'] = new_df.apply(fillNegative, axis=1)
```

```python
new_df = new_df.drop(labels=['retweet_count','sentiment'],axis=1)
```

```python
new_df = new_df.groupby('created_at').sum()
```

```python
# Export comma-separated variable file (just in case)
new_df.to_csv('DATE-SENTIMENT.csv', index=True)
```
